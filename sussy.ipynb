```python
import easygopigo3 as easy
import time
gpg = easy.EasyGoPiGo3()
ds = gpg.init_distance_sensor()
disServ = gpg.init_servo(1)
```

```python
class Servo:
    def __init__(self, port, zeroLocation, timePerDegree):
        self.zeroLocation = zeroLocation
        self.gpgServo = gpg.init_servo(port)
        self.gpgServo.rotate_servo(zeroLocation)
        self.currentLocation = 0
        self.timePerDegree = timePerDegree

    def centerServo(self):
        self.gpgServo.rotate_servo(self.zeroLocation)
        self.currentLocation = 0

    def rotateServo(self, degrees):
        self.gpgServo.rotate_servo(self.zeroLocation + degrees)
        self.currentLocation = degrees

    def rotateServoRelative(self, degrees):
        self.rotateServo(self.currentLocation + degrees)

distServo = Servo(1, 69, 0.005)
```

```python
def rayCastIRL():
    #scanDirection = distServo.currentLocation % -1
    distServo.rotateServo(-60)
    curAngle = -60
    minAngle = -60
    minDist = 10000
    time.sleep(0.3)
    
    while (curAngle <= 60):
        ds.rotateServoRelative(5)
        curAngle += 5
        time.sleep(distServo.timePerDegree)
        if(ds.read_mm < minDist):
            minDist = ds.read_mm
            minAngle = curAngle
            
    gpg.turn_degrees(minAngle)
    rayCastIRL()
```

```python
def avoid():
    """
    distance = ds.read_mm()
    if (distance < 100):
        gpg.stop()
        #rayCastIRL()
        
    else:
        gpg.forward()
    """
    gpg.forward()
    rayCastIRL(ds)
```

```python
def dontdie():
    while True:
        avoid()

dontdie()
```
